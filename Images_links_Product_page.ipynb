```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

# Define a function to scrape image URLs from a specific product page
def scrape_image_urls(product_url):
    try:
        r = requests.get(product_url)
        r.raise_for_status()  # Raise an exception for HTTP errors

        soup = BeautifulSoup(r.text, "html.parser")

        # Find all image tags with a specific class (replace 'q6DClP' with the actual class name)
        image_tags = soup.find_all('img', class_='q6DClP')

        # Extract and store the image links in a list
        image_links = [img['src'] for img in image_tags]

        return image_links
    except requests.exceptions.RequestException as e:
        print(f"Request Error: {e}")
        return []
    except Exception as ex:
        print(f"Error: {ex}")
        return []

# Read the CSV file with product URLs
csv_path = "flipkart_href_all(New).csv"  # Update with the actual path to your CSV file
df = pd.read_csv(csv_path)

# Create a new DataFrame to store the scraped image URLs
image_urls_df = pd.DataFrame(columns=["Product_URL", "Image_1", "Image_2", "Image_3", "Image_4", "Image_5", "Image_6", "Image_7", "Image_8", "Image_9", "Image_10"])

# Iterate over the product URLs and scrape image URLs
for index, row in df.iterrows():
    product_url = row["Href"]
    if not pd.isna(product_url):
        # Construct the complete URL by adding "https://www.flipkart.com/"
        product_url = urljoin("https://www.flipkart.com/", product_url)
        image_links = scrape_image_urls(product_url)

        if image_links:
            # Create a dictionary with product URL and image URLs
            image_data = {"Product_URL": product_url}
            for i, link in enumerate(image_links[:10]):  # Limit to 10 images per product
                image_data[f"Image_{i + 1}"] = link

            # Append the scraped data as a new row to the DataFrame
            image_urls_df = image_urls_df.append(image_data, ignore_index=True)

# Save the scraped image URLs to a new CSV file
output_csv_path = "D:/flipkart_image_links_final.csv"  # Update with the desired output path
image_urls_df.to_csv(output_csv_path, index=False)

print(f"Image URLs saved to {output_csv_path}")
```
